## 3. Reshaping

```python
arr = np.arange(10, dtype=float).reshape((2, 5))
print(arr)
print(arr.shape)
```

```python
print(arr.reshape(5, 2))
```

### Add an index

```python
a = np.array([0, 1])
a.shape
```

```python
a_col = a[:, np.newaxis]
print(a_col)
a_col.shape
```

```python
a_col = a[:, None]
print(a_col)
```

### Transpose

```python
print(a_col.T)
```

### Flatten: 항상 원 배열의 평활화된 복사본(copy)을 반환한다

```python
arr_flt = arr.flatten()
arr_flt[0] = 33
print(arr_flt)
print(arr)
```

### Ravel: 가능한 한 원 배열의 뷰(view)를 반환한다
* 큰 배열로 작업할 때는 메모리를 적게 사용하므로 프로그램 수행이 빠를 수 있으나
* view 를 수정하면 원 배열도 수정되므로 주의를 요함

```python
arr_flt = arr.ravel()
arr_flt[0] = 33
print(arr_flt)
print(arr)
```

```python
arr_flt = arr.ravel(order='F')    # The same order as FORTRAN (default is 'C')
arr_flt[0] = 33
print(arr_flt)
print(arr)
```

## 6. Slicing

**Syntax**: start:stop:step  
__with__ start _(default 0)_, stop _(default last)_, step _(default 1)_

```python
arr = np.arange(10, dtype=float).reshape((2, 5))
arr
```

```python
arr[0, :]           # row 0: returns 1d array
```

```python
arr[:, 0]           # column 0: returns 1d array
```

```python
arr[:, :2]          # columns strictly before index 2 (2 first columns)
```

```python
arr[:, 2:]          # columns after index 2 included
```

```python
arr2 = arr[:, 1:4]  # columns between index 1 (included) and 4 (excluded)
print(arr2)
```

### Slicing returns a view (not a copy)

```python
arr2[0, 0] = 33
print(arr2)
print(arr)
```

### Reverse order

```python
arr = np.arange(20, dtype=float).reshape((4, 5))
arr
```

```python
print(arr[::-1])
```

```python
print(arr[1 ::-1])
```

```python
print(arr[1, ::-1])
```

## 7. Fancy indexing: Integer or boolean array indexing
* **Fancy indexing returns a copy not a view.**

### Integer array indexing

```python
arr2 = arr[:, [1,2,3]]    # return a copy
print(arr2)
```

```python
arr2[0, 0] = 44
print(arr2)
print(arr)
```

### Boolean arrays indexing

```python
arr2 = arr[arr > 4]     # return a copy
print(arr2)
```

```python
arr2[0] = 44
print(arr2)
print(arr)
```

```python
# lvalue indexing에 의한 할당에서는 값을 변경시킨다.
arr[arr > 5] = 0
print(arr)
```

### More Boolean arrays indexing

```python
names = np.array(['Bob', 'Joe', 'Will', 'Bob'])
```

```python
names == 'Bob'                        # returns a boolean array
```

```python
names[names != 'Bob']                 # logical selection
```

```python
(names == 'Bob') | (names == 'Will')  # keywords "and/or" don't work with boolean arrays
```

```python
names[names != 'Bob'] = 'Joe'         # assign based on a logical selection
names
```

```python
np.unique(names) # set function
```

## 8. Vectorized operations

```python
nums = np.arange(5)
nums * 10            # multiply each element by 10
```

```python
nums = np.sqrt(nums) # square root of each element
nums
```

```python
np.ceil(nums)        # also floor, rint (round to nearest int)
```

```python
np.isnan(nums)       # checks for NaN
```

```python
nums + np.arange(5)  # add element-wise
```

```python
np.maximum(nums, np.array([1, -2, 3, -4, 5]))    # compare element-wise
```

```python
# Compute Euclidean distance between 2 vectors
vec1 = np.random.randn(10)
vec2 = np.random.randn(10)
dist = np.sqrt(np.sum((vec1 - vec2) ** 2))
dist
```

```python
# math and stats
rnd = np.random.randn(4, 2) # random normals in 4x2 array
print(rnd)
print(rnd.mean())
print(rnd.std())
```

```python
print(rnd.argmin())     # index of minimum element
print(rnd.sum())
print(rnd.sum(axis=0))  # sum of columns
print(rnd.sum(axis=1))  # sum of rows
```

```python
# methods for boolean arrays
(rnd > 0).sum() # counts number of positive values
```

```python
(rnd > 0).any() # checks if any value is True
```

```python
(rnd > 0).all() # checks if all values are True
```

```python
# random numbers
np.random.seed(12234)       # Set the seed
```

```python
np.random.rand(2, 3)        # 2 x 3 matrix in [0, 1]
```

```python
np.random.randn(10)         # random normals (mean 0, sd 1)
```

```python
np.random.randint(0, 2, 10) # 10 randomly picked 0 or 1
```

## 9. Broadcasting  
다른 크기의 배열들 간의 연산을 가능하게 하는 암묵적인 변환을 의미한다. 즉, 작은 배열을 늘리거나 브로드캐스트해서 배열의 크기가 큰 배열에 맞도록 만든다.

```python
a = np.array([[ 0, 0, 0],
            [10, 10, 10],
            [20, 20, 20],
            [30, 30, 30]])
b = np.array([0, 1, 2])
print(a + b)
```

**A** (2d array): 5 x 4  
**B** (1d array): 1  
_Result_ (2d array): 5 x 4  

**A** (2d array): 5 x 4  
**B** (1d array): 4  
_Result_ (2d array): 5 x 4  

**A** (3d array): 15 x 3 x 5  
**B** (3d array): 15 x 1 x 5  
_Result_ (3d array): 15 x 3 x 5  

**A** (3d array): 15 x 3 x 5  
**B** (2d array): 3 x 5  
_Result_ (3d array): 15 x 3 x 5  

**A** (3d array): 15 x 3 x 5  
**B** (2d array): 3 x 1  
_Result_ (3d array): 15 x 3 x 5

## 10. Save and Load Arrays

* Numpy array를 파일로 저장했다가 다시 로드하여 사용하는 방법.
* 기계학습의 학습 과정 중간이나 학습 완료 후 파라미터를 저장했다가 나중에 다시 로드하여 사용할 수 있다.
* 기계학습 알고리즘이 사용하기 용이하도록 데이터셋을 numpy array로 저장할 수 있다.

### np.save, np.load: 1 개의 array를 파일로 저장 및 로드 
* 파일 확장자: `.npy`

```python
a = np.array([[ 0, 0, 0],
            [10, 10, 10],
            [20, 20, 20],
            [30, 30, 30]])
b = np.array([0, 1, 2])
```

```python
np.save('.\\numpy_array', a)
```

```python
a_loaded = np.load('.\\numpy_array.npy')
a_loaded
```

### np.savez, np.load: 여러개의 array를 1 개의 비압축 파일로 저장 및 로드
* 파일 확장자: `.npz`

```python
np.savez('.\\numpy_array_multi', a, b)
```

```python
saved_a_b = np.load('.\\numpy_array_multi.npz')
type(saved_a_b)
```

```python
saved_a_b.files
```

```python
a_loaded = saved_a_b['arr_0']
b_loaded = saved_a_b['arr_1']
a_loaded, b_loaded
```

```python
saved_a_b.close()
```

#### array name 지정하기

```python
np.savez('.\\numpy_array_multi', a_saved=a, b_saved=b)
saved_a_b = np.load('.\\numpy_array_multi.npz')
saved_a_b.files
```

```python
a_loaded = saved_a_b['a_saved']
b_loaded = saved_a_b['b_saved']
saved_a_b.close()
a_loaded, b_loaded
```
