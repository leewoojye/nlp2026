```python
from io import StringIO

from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfparser import PDFParser

fake_file_handle = StringIO()
with open(sample_pdf, 'rb') as in_file:
    parser = PDFParser(in_file)
    doc = PDFDocument(parser)
    rsrcmgr = PDFResourceManager()
    device = TextConverter(rsrcmgr, fake_file_handle, laparams=LAParams())
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    for page in PDFPage.create_pages(doc):
        interpreter.process_page(page)

print(fake_file_handle.getvalue())
```

### 텍스트 추출 -- All

```python
from io import StringIO
from pdfminer.converter import TextConverter
from pdfminer.pdfinterp import PDFPageInterpreter
from pdfminer.pdfinterp import PDFResourceManager
from pdfminer.pdfpage import PDFPage
 
def extract_text_from_pdf_once(pdf_path):
    resource_manager = PDFResourceManager()
    fake_file_handle = StringIO()
    converter = TextConverter(resource_manager, fake_file_handle) 
    page_interpreter = PDFPageInterpreter(resource_manager, converter)
 
    with open(pdf_path, 'rb') as f:
        for page in PDFPage.get_pages(f, caching=True, check_extractable=True):
            page_interpreter.process_page(page)
        text = fake_file_handle.getvalue()
 
    converter.close()
    fake_file_handle.close()
 
    if text: 
        return text

print(extract_text_from_pdf_once(sample_pdf))
```

###  텍스트 추출 -- Pages

```python
from io import StringIO
from pdfminer.converter import TextConverter
from pdfminer.pdfinterp import PDFPageInterpreter, PDFResourceManager
from pdfminer.pdfpage import PDFPage
 
def extract_text_by_page(pdf_path):
    with open(pdf_path, 'rb') as f:
        for page in PDFPage.get_pages(f, caching=True, check_extractable=True):
            resource_manager = PDFResourceManager()
            fake_file_handle = StringIO()
            converter = TextConverter(resource_manager, fake_file_handle)
            page_interpreter = PDFPageInterpreter(resource_manager, converter)
            page_interpreter.process_page(page)
 
            text = fake_file_handle.getvalue()
            yield text
 
            converter.close()
            fake_file_handle.close()
    return
 
def extract_text_from_pdf(pdf_path):
    for page in extract_text_by_page(pdf_path):
        print(page)
        print()
    return

print(extract_text_from_pdf(sample_pdf))
```

### PDF to XML

```python
import os
import xml.etree.ElementTree as xml
from xml.dom import minidom
 
def export_as_xml(pdf_path, xml_path):
    filename = os.path.splitext(os.path.basename(pdf_path))[0]
    root = xml.Element('{filename}'.format(filename=filename))
    pages = xml.Element('Pages')
    root.append(pages)
 
    counter = 1
    for page in extract_text_by_page(pdf_path):
        text = xml.SubElement(pages, 'Page_{}'.format(counter))
        text.text = page[0:100]
        counter += 1
 
    tree = xml.ElementTree(root)
    xml_string = xml.tostring(root, 'utf-8')
    parsed_string = minidom.parseString(xml_string)
    pretty_string = parsed_string.toprettyxml(indent='  ')
 
    with open(xml_path, 'w', encoding='utf-8') as f:
        f.write(pretty_string)
    return

export_as_xml(sample_pdf, Dataset_Path+'out.xml')
viewfile(Dataset_Path+'out.xml')
```

### PDF to JSON

```python
import json
import os
 
def export_as_json(pdf_path, json_path):
    filename = os.path.splitext(os.path.basename(pdf_path))[0]
    data = {'Filename': filename}
    data['Pages'] = []
 
    counter = 1
    for page in extract_text_by_page(pdf_path):
        text = page[0:100]
        page = {'Page_{}'.format(counter): text}
        data['Pages'].append(page)
        counter += 1
 
    with open(json_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False)
    return

export_as_json(sample_pdf, Dataset_Path+'out.json')
viewfile(Dataset_Path+'out.json')
```

### PDF to CSV

```python
import csv
import os
 
def export_as_csv(pdf_path, csv_path):
    filename = os.path.splitext(os.path.basename(pdf_path))[0]
 
    counter = 1
    with open(csv_path, 'w', encoding='utf-8') as csv_file:
        writer = csv.writer(csv_file)
        for page in extract_text_by_page(pdf_path):
            text = page[0:100]
            words = text.split()
            writer.writerow(words)
    return
 
export_as_csv(sample_pdf, Dataset_Path+'out.csv')
viewfile(Dataset_Path+'out.csv')
```

# Converting a PDF file to text

* 대부분의 PDF 파일은 잘 구성된 텍스트를 포함하고 있는 것처럼 보인다. 그러나 현실은 PDF 파일에는 단락, 문장 또는 단어와 유사한 내용이 들어 있지 않다. 텍스트에 관한 한 PDF 파일은 문자와 그 배치만을 인식한다.
* 이러한 것들이 PDF 파일에서 의미있는 텍스트 조각들을 추출하기 어렵게 만든다. 단락을 구성하는 문자는 표, 페이지 바닥글 또는 그림의 설명을 구성하는 문자와 다르지 않다. .txt 파일이나 워드 문서와 같은 여타 문서 형식과는 달리, PDF 형식은 텍스트 스트림을 포함하지 않는다.
* PDF 문서는 하나 이상의 페이지의 모양을 함께 기술하는 객체 모음으로 구성되며, 추가 대화형 요소와 상위 수준의 응용프로그램 데이터와 함께 제공될 수 있다. PDF 파일은 관련된 구조 정보와 함께 PDF 문서를 구성하는 개체를 포함하며, 모두 단일 자체 포함 바이트 시퀀스로 표현된다.

## Layout analysis algorithm

* PDFMiner는 문자 위치에 대한 휴리스틱스를 사용하여 이러한 구조 중 일부를 재구성하려고 한다. 이것은 문장과 단락에 효과적이다. 왜냐하면 가까운 문자들의 의미있는 그룹이 만들어질 수 있기 때문이다.
* 레이아웃 분석은 세 가지 단계로 구성된다. 문자를 단어와 선으로 그룹화한 다음 선들을 상자로 그룹화하고 마지막으로 텍스트상자를 계층적으로 그룹화한다. 이러한 단계는 다음 절에서 논의한다. 배치 분석의 결과 출력은 PDF 페이지에 있는 배치 객체의 순서 계층이다.



[The output of the layout analysis is a hierarchy of layout objects.]

* 배치 분석의 출력은 몇 가지 매개변수들에 따라 크게 달라진다. 이 모든 매개변수는 LAParams 클래스의 일부분이다.

### Grouping characters into words and lines
* 문자에서 텍스트로 가는 첫 번째 단계는 의미있는 방식으로 문자들을 그룹화하는 것이다. 각 문자는 왼쪽 하단 모서리와 오른쪽 상단 모서리에 대해 x 좌표와 y 좌표를 가지고 있다. Pdfminer.six는 이 경계 상자(bounding box)를 사용하여 어떤 문자들이 함께 속하는지 결정한다.
* 가로와 세로로 모두 가까운 문자들끼리 그룹화한다. 얼마나 가까워야 하는지는 char_margin(그림에서 M)과 line_overlap(그림에는 없음) 매개변수에 의해 결정된다. 두 문자의 경계 상자 사이의 수평 거리는 char_margin과 경계 상자 사이의 수직 중첩이 line_overlap보다 작아야 한다.



* char_margin과 line_overlap의 값은 문자의 경계 상자의 크기에 상대적이다. char_margin은 경계 상자 중 하나의 최대 폭에 상대적이며, line_overlap은 경계 상자 중 하나의 최소 높이에 상대적이다.
* PDF 형식은 공백 문자의 개념이 없기 때문에 문자 사이에 공간을 삽입해야 한다. 글자가 word_margin(그림에서 W)보다 멀리 떨어져 있으면 공백이 삽입된다. 단어_margin은 새 문자의 최대 너비 또는 높이에 상대적이다. 더 큰 단어_margin을 갖는 것은 더 작은 단어들을 만들고 더 자주 글자 사이에 공간을 삽입한다. 단어_margin은 char_margin보다 작아야 하며 그렇지 않으면 모든 문자가 공백으로 분리된다.
* 이 무대의 결과는 선 목록이다. 각 행은 등장인물의 목록을 구성한다. 이러한 문자는 PDF 파일에서 유래한 원본 LTChar 문자 또는 각 행의 끝에 단어 또는 새 줄 사이의 공백을 나타내는 LTAnno 문자를 삽입했다.


### Grouping lines into boxes
* 두 번째 단계는 의미 있는 방식으로 라인을 그룹화하는 것이다. 각 선에는 포함된 문자의 경계 상자에 의해 결정되는 경계 상자가 있다. 그룹화 캐릭터처럼 pdfminer.6은 경계 상자를 사용하여 선을 그룹화한다.
* 수평으로 겹치는 선과 수직으로 가까운 선이 모두 그룹화된다. 라인을 수직으로 얼마나 닫아야 하는지는 line_margin에 의해 결정된다. 이 여유는 경계 상자의 높이에 비례하여 지정된다. 경계 상자의 상단(그림의 L1 참조)과 하단(그림의 L2 참조) 사이의 간격이 절대선 여백보다 더 가까울 경우, 즉 경계 상자의 높이에 곱한 선_마진(line_margin)이 가깝다면 선이 가깝다.



* 이 단계의 결과는 텍스트 상자 목록이다. 각 상자는 선 목록으로 구성되어 있다.

### Grouping textboxes hierarchically
* 마지막 단계는 의미 있는 방법으로 텍스트 상자를 그룹화하는 것이다. 이 단계는 서로 가장 가까운 두 개의 텍스트 상자를 반복적으로 병합한다.
* 경계 상자의 근접도는 두 텍스트 상자 사이의 영역(그림의 파란색 영역)으로 계산된다. 즉, 개별 선의 경계 상자의 면적을 뺀, 두 선을 모두 둘러싸고 있는 경계 상자의 면적이다.





## Working with rotated characters
* 위에서 설명한 알고리즘은 모든 문자의 방향이 동일하다고 가정한다. 그러나 PDF에서는 어떤 쓰기 방향도 가능하다. 이를 수용하기 위해 pdfminer.6는 detect_수직 매개변수를 사용하여 수직 쓰기를 탐지할 수 있다. 이렇게 하면 PDF가 90도(또는 270도) 회전한 것처럼 모든 그룹화 단계가 적용된다.
