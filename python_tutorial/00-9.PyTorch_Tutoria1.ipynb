**텐서**를 사용하는 이유 중 하나는 *벡터화 연산*이다: 텐서의 특정 차원에 대해 병렬로 수행할 수 있는 연산이다.

- PyTorch에서 sum(dim=n)과 같은 연산을 사용할 때, dim 인자는 어떤 차원을 따라 연산을 수행할지를 지정한다.
- 여기서 '따라 연산을 수행한다'는 것은 해당 차원이 줄어들거나(collapsed) 합쳐진다는 의미.
  - dim=0 : 행(row)
  - dim=1 : 열(column)

```python
import torch

# 2x3 텐서 생성
data_2d = torch.arange(1, 36, dtype=torch.float32).reshape(5, 7)
print("원본 텐서:\n", data_2d)
print("원본 텐서의 shape:", data_2d.shape)

# dim=0 (행을 따라 합계): 각 열의 합을 계산
sum_dim0 = data_2d.sum(dim=0)
print("\ndim=0으로 합계 (각 열의 합):", sum_dim0)
print("sum_dim0의 shape:", sum_dim0.shape)

# dim=1 (열을 따라 합계): 각 행의 합을 계산
sum_dim1 = data_2d.sum(dim=1)
print("\ndim=1으로 합계 (각 행의 합):", sum_dim1)
print("sum_dim1의 shape:", sum_dim1.shape)

# 각 행의 표준편차 계산 -> dim=1
print("행별 표준편차:") # 행의 표준편차
std_dim1 = data_2d.std(dim=1)
print("\ndim=1으로 표준편차 (각 행의 표준편차):",std_dim1)
print("std_dim1의 shape:", std_dim1.shape)
```

```python
data = torch.arange(1, 7, dtype=torch.float32).reshape(1, 2, 3) # 배치 크기기 , 2행3열
print(data)
print()
print(data.sum(dim=0))
print(data.sum(dim=0).shape)
print()
print(data.sum(dim=0).sum(dim=0))
print(data.sum(dim=0).sum(dim=0).shape)
print()
print(data.sum(dim=0).sum(dim=1))
print(data.sum(dim=0).sum(dim=1).shape)
```

## 자동 미분 (Autograd)
Pytorch는 자동 미분 기능으로 잘 알려져 있다. `backward()` 메서드를 호출하여 PyTorch가 **기울기**(Gradient)를 계산하도록 요청할 수 있으며, 계산된 기울기는 `grad` 속성에 저장된다.

```python
# 예제 텐서 생성
# requires_grad 매개변수는 PyTorch에게 기울기를 저장하도록 지시한다
x = torch.tensor([2.], requires_grad=True)

# 기울기가 계산되면 기울기를 프린트
# 지금은 x가 스칼라이므로 `None`이 출력된다.
print(x.grad)
```

```python
# x에 대한 y의 기울기 계산
y = x * x * 3  # 3x^2
y.backward()
print(x.grad)  # d(y)/d(x) = d(3x^2)/d(x) = 6x = 12
```

다른 텐서에서 다시 역전파를 실행하여 어떤 일이 발생하는지 확인해 보자.

```python
z = x * x * 3  # 3x^2
z.backward()
print(x.grad)
```

```python
x.grad = None
z = x * x * 3  # 3x^2
z.backward()
# y = x * x * 3
print(x.grad)
```

```python
z = x * x * 3  # 3x^2
z.backward()
# y = x * x * 3
print(x.grad)
```

```python
z = x * x * 3  # 3x^2
z.backward()
# y = x * x * 3
print(x.grad)
```

### 신경망 학습 시 `옵티마이저.zero_grad()`가 꼭 필요한 이유
- `x.grad`는 지금까지 계산된 그래디언트들의 합으로 업데이트됨
- `.backward()`를 여러 번 호출하면 그 결과가 누적됨. 역전파의 기본구조 때문에 신경망에서는 하나의 뉴런에 여러 경로를 통해 전달된 기울기를 모두 더한 후 가중치를 업데이트
- 매번 학습할 때마다 옵티마이저의 `zero_grad()` 메서드를 호출해서 이전 기울기 값을 초기화해야 함 그렇지 않으면 기울기 누적으로 인해 잘못된 방향으로 학습될 수 있음

### **Linear Layer**
### Linear Layer (선형 계층)

**선형 계층**(Linear layer)은 기본적으로 **행렬 곱셈**과 **편향(bias) 덧셈**을 수행하는 연산을 정의한다.  
즉, 입력 `x`에 대해 아래 연산을 실행한다:

\[
Ax + b
\]

여기서:  
- `A`는 **가중치 행렬 (weight matrix)**  
- `b`는 **편향 벡터 (bias vector)**  

PyTorch에서는 `nn.Linear`를 생성할 때, **가중치 `A`와 편향 `b`를 무작위(random)** 값으로 초기화한다.

---

### 입력 차원

`nn.Linear(H_i, H_o)`는 입력의 **마지막 차원**이 `H_i`인 데이터를 받아서, **마지막 차원을 `H_o`로 변환**한다.

입력 텐서는 보통 다음과 같은 형태를 가진다:

\[
(N, *, H_i)
\]

- `N`: 일반적으로 **batch size**를 의미  
- `*`: 중간에 올 수 있는 임의의 다른 차원들. 이들은 선형 변환 전에 **flatten**되지 않고 그대로 유지된다.  
- `H_i`: **입력 벡터의 크기**, 즉 마지막 차원의 크기


```python
# 입력 생성
input = torch.ones(2,3,4)
print(input)
print()
# (N,*,H_in) 차원의 입력을 (N,*,H_out) 차원의 출력으로 변환하는 선형 레이어 생성
linear = nn.Linear(4, 2)
linear_output = linear(input)
linear_output
```

```python
linear_output.shape
```

```python
list(linear.parameters()) # Ax + b
```

- 입력 = (batch_size, input_dim) # 4
- 출력 = (batch_size, output_dim) # 2
- (input_dim, output_dim) 형태의 선형 레이어

### **Putting the Layers Together**
- 지금까지는 레이어를 만들고, 한 레이어의 출력을 다음 레이어의 입력으로 넘기는 방식을 했음
- 그런데 이렇게 하면 중간 텐서를 계속 생성하고 넘겨줘야 해서 번거로울 수 있음
- 그래서 nn.sequential을 사용하여 그 과정을 자동화 처리함

```python
block = nn.Sequential(
    nn.Linear(4, 2),
    nn.Sigmoid()
)

input = torch.ones(2,3,4) # 배치 크기, 시퀀스 길이, 특성
output = block(input)
output
```

### 사용자 정의 모듈 만들기 (Custom Module)

PyTorch에서는 `nn.Linear`, `nn.Conv2d`와 같은 **미리 정의된 모듈** 외에도,  
직접 `nn.Module`을 상속받아 **나만의 모듈(Custom Module)**을 만들 수 있다.  
이 방식은 **복잡한 구조의 모델**을 설계할 때 매우 유용하다.

---

#### 사용자 정의 모듈을 만들기 위한 구성 요소

1. `nn.Module`을 상속  
2. `__init__()` 메서드에서 필요한 구성 요소(레이어 등) 초기화  
3. `forward(x)` 메서드를 **반드시 정의**해야 함

---

#### 학습 가능한 파라미터 자동 등록

`nn.Module`의 속성으로 정의된 다음과 같은 요소들은  
**자동으로 학습 대상 파라미터**로 등록되어 `model.parameters()`에 포함된다:

- `nn.Linear`, `nn.Conv2d`  
- `nn.Parameter`  

즉, 따로 등록할 필요 없이 학습 대상이 된다.

---

#### `nn.Sequential` vs 사용자 정의 모듈

- `nn.Sequential`: 간단하고 빠르게 레이어를 쌓을 때 유용  
- 사용자 정의 모듈: **더 유연한 모델 구성**이 가능  
  - 조건문, 반복문, 병렬 연산 등 복잡한 로직 구현이 필요할 때 적합



## Multi-Layer Perceptron (MLP)

#### [방법-1] `torch.nn.Sequential` 사용

```python
class MultilayerPerceptron(nn.Module):

  def __init__(self, input_size, hidden_size):
    # super class의 __init__ 함수 호출
    super(MultilayerPerceptron, self).__init__() # Initializer 호출을 통해 학습가능한 파라미터 등록,모델 상태 관리,GPU 전송 등 기능 작동

    # 주요 기본 정보 보관: 초기화 파라미터 저장
    self.input_size = input_size
    self.hidden_size = hidden_size

    # 모델 정의
    self.model = nn.Sequential(
        nn.Linear(self.input_size, self.hidden_size), # 입력 -> 은닉층
        nn.ReLU(), # 비선형성
        nn.Linear(self.hidden_size, self.input_size), # 은닉층 -> 출력
        nn.Sigmoid() # 출력에 시그모이드
    )

  def forward(self, x):
    output = self.model(x)
    return output
```

#### [방법-2] `__init__` 메서드에서 개별 층을 직접 정의하고 `forward`에서 층들을 연결하는 방법

```python
class MultilayerPerceptron(nn.Module):

  def __init__(self, input_size, hidden_size):
    # Call to the __init__ function of the super class
    super(MultilayerPerceptron, self).__init__()

    # Bookkeeping: Saving the initialization parameters
    self.input_size = input_size
    self.hidden_size = hidden_size

    # Defining of our layers
    self.linear = nn.Linear(self.input_size, self.hidden_size)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(self.hidden_size, self.input_size)
    self.sigmoid = nn.Sigmoid()

  def forward(self, x):
    linear = self.linear(x)
    relu = self.relu(linear)
    linear2 = self.linear2(relu)
    output = self.sigmoid(linear2)
    return output
```

이제 클래스를 정의했으니, 인스턴스를 생성해서 어떤 동작을 하는지 살펴볼 수 있다.

```python
# 입력 데이터 준비
input = torch.randn(2, 5) #배치 수, 특성 수

# 모델 생성
model = MultilayerPerceptron(5, 3) # (input_size, hidden_size)

# 모델에 입력하기
model(input)
```

`named_parameters()` 및 `parameters()` 메서드를 사용하여 모델의 매개변수를 검사할 수 있다.

```python
list(model.named_parameters())
```

### 옵티마이저(Optimizer)의 역할

- 역전파로 **기울기**(gradient)만 계산된다고 해서 학습이 끝나는 것이 아니다.  
- 그 **기울기를 바탕으로 실제로 파라미터를 업데이트**하는 것이 바로 **optimizer**(최적화 도구)의 역할이다.

---

### PyTorch 옵티마이저 사용 흐름

1. **옵티마이저 선택**  
   - 예: `SGD`, `Adam` 등

2. **모델의 파라미터 전달**  
   - `model.parameters()`를 통해 전달

3. **학습 루프 내에서 반복 수행**  
   - `optimizer.zero_grad()`  
     → 이전 step의 기울기를 초기화  
   - `loss.backward()`  
     → 현재 step의 **기울기 계산 (역전파)**  
   - `optimizer.step()`  
     → 계산된 기울기를 사용하여 **파라미터 업데이트**


```python
import torch.optim as optim
```

### 손실 함수(Loss Function) 정의

옵티마이저를 생성한 후에는,  
어떤 **손실(loss)**을 줄이고 싶은지 **정의**해야 한다.

손실 함수를 정의하는 방법:

1. **손실 함수를 직접 정의**하기  
2. **PyTorch에서 제공하는 손실 함수** 사용  
   - 예: `nn.BCELoss()`, `nn.MSELoss()` 등


```python
# 목표값 y 데이터 만들기
y = torch.ones(10, 5)

# 목표값 y에 노이즈를 첨가하여 우리가 사용할 x 생성
x = y + torch.randn_like(y)
x
```

이제, Model, Optimizer 및 Loss Function을 정의할 수 있다.

```python
# 모델 인스턴스화
model = MultilayerPerceptron(5, 3)

# optimizer 정의
adam = optim.Adam(model.parameters(), lr=1e-1) # 여기서 학습률 lr=0.1 은 비교적 큰 값이다

# 사전에 정해진 손실함수을 사용하여 손실을 정의
loss_function = nn.MSELoss()

# 모델의 현재 학습 상태를 계산
y_pred = model(x)
loss_function(y_pred, y).item()
```

- 이제 모델의 손실이 더 작아지고 있는지(학습이 잘 되고 있는지) 확인해 보자.
- 필요한 모든 것이 준비되었으니 학습 루프를 설정할 수 있다.

```python
# 훈련 반복 횟수를 나타내는 epoch 수 설정
n_epoch = 10

for epoch in range(n_epoch):

  adam.zero_grad() # 기울기 계산은 누적되므로 gradients = 0 로 클리어

  # 모델이 산출하는 예측값을 얻는다
  y_pred = model(x)

  # 손실 함수값 계산
  loss = loss_function(y_pred, y)

  # 상태(stats) 출력
  print(f"Epoch {epoch}: traing loss: {loss}")

  # 기울기 계산
  loss.backward()

  # 가중치 최적화를 위해 훈련 진행
  adam.step()
```

```python
list(model.parameters())
```

You can see that our loss is decreasing. Let's check the predictions of our model now and see if they are close to our original `y`, which was all `1s`.

- 손실이 감소하는 것을 볼 수 있다.
- 이제 모델의 예측값을 확인하고 초기값인 'y'(모두 1)에 얼마나 가까운지 살펴보자.

```python
# See how our model performs on the training data
y_pred = model(x)
y_pred
```
